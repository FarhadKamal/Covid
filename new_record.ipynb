```python
import pandas as pd

df= pd.read_csv("covid.csv",parse_dates=['Date'])

print("New Cases")
new_cases=int(input())
print("New Deaths")
new_deaths=int(input())
print("New Recoveries")
new_recoveries=int(input())


total_cases=df['Total Cases'].max()+new_cases
total_deaths=df['Total Deaths'].max()+new_deaths
total_recoveries=df['Total Recoveries'].max()+new_recoveries
new_date=df['Date'].max()+ pd.DateOffset(1)

new_df= pd.DataFrame({
        "Date": [new_date],
        "Total Cases": [total_cases],
        "New Cases": [new_cases],
        "Total Deaths": [total_deaths],
        "New Deaths": [new_deaths],
        "Total Recoveries": [total_recoveries],
        "New Recoveries": [new_recoveries]

})

final_df = pd.concat([df,new_df], ignore_index=True)
final_df.tail()

final_df.to_csv("covid.csv",index=False)
final_df.tail()
```
